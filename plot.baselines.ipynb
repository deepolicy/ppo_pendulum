```python
import re
from matplotlib import pyplot as plt
```

```python
file = '/media/inksci/file/deepolicy/gym-envs/useless/baselines-2/print.txt'
```

```python
with open(file, 'r') as f:
    _str = f.read()
    print(_str[:3000])
    
```

```python
_str.split('\n')[:10]
```

```python
_dict = {}
for line in _str.split('\n'):
    match_obj = re.match(r'\| (.*)\s+\| (.*)\s+\|', line, re.M | re.I)
    if match_obj:
        
        k, v = match_obj.group(1), float(match_obj.group(2))
        
        if not k in _dict:
            _dict[k] = []
            
        _dict[k].append(v)
```

```python
for k in _dict:
    plt.title(k)
    plt.plot(_dict[k])
    plt.show()
```
